```python
def model(X, w, b):
    return ((X @ w) + b)
```

```python
def cost_func(X, y, w, b):
    m, n = X.shape

    error = model(X, w, b) - y
    mean_squared_error = np.sum((error ** 2)) / m
    return mean_squared_error
```

```python
def gradient_func(X, y, w, b):
    m, n = X.shape
    dj_dw = np.zeros((n,))
    dj_db = 0.0

    error = model(X, w, b) - y
    dj_dw = 2 * (X.T @ error) / m
    dj_db = 2 * np.sum(error) / m
    return dj_dw, dj_db
```

```python
def learning_schedule(t):
    return (t0 / (t + t1))
```

```python
def batch_gradient_descent(X, y, w, b, epoch):
    m, n = X.shape
    cost_history = []

    for e in range(epoch):
        dj_dw, dj_db = gradient_func(X, y, w, b)
        alpha = learning_schedule(epoch * m + e)

        w = w - alpha * dj_dw
        b = b - alpha * dj_db

        cost_history.append(cost_func(X, y, w, b))

        if e % 100 == 0:
            print(f"Epoch: {e:4d} | Loss: {cost_history[-1]:8.2f}")
    return w, b, cost_history
```

```python
X = np.random.random((500,5))
weights = np.array([6., 3., 4.5, 3., 2.5])
y = X.dot(weights) + 4 + np.random.random((500,))
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
m, n = X_scaled.shape

epoch = 15000
w_init = np.zeros((n,))
b_init = 0.0
t0, t1 = 1500, 15000 
w, b, cost_history = batch_gradient_descent(X_scaled, y, w_init, b_init, epoch)
```

```python
print("Weights:", w, "Bias:", b)
```

```python
from sklearn.linear_model import SGDRegressor, LinearRegression

sgd_reg, lin_reg = SGDRegressor().fit(X_scaled, y), LinearRegression().fit(X_scaled, y)
print("SGD Weights:", sgd_reg.coef_, "SGD Bias:", sgd_reg.intercept_)
print("LR Weights:", lin_reg.coef_, "LR Bias:", lin_reg.intercept_)
```
